## 2. Combine one N and one T after filtering re

```r
INDIR="/data2/kantian/LineageTracing/SMALT/0.Results/2.final_phy_re/"
OUTDIR="/data2/kantian/LineageTracing/SMALT/0.Results/2.filteredNT_re/"
if (! file.exists(OUTDIR)){
        dir.create(file.path(OUTDIR))
}
setwd(INDIR)

Ns=c("2_N","2_N","4_N","5_N","16_N","19_N","19_N","19_N","19_N","47_N1","47_N4","47_N5","47_N6","47_N8",
    "49_N","49_N","50_N","65_N","65_N","66_N",
     "132_N","151_N","17_N","18_N","18_N","142_N","142_N","142_N","148_N","148_N",
     "68_N","68_N","68_N","68_N","68_N","68_N","68_N","68_N","68_N",#"49_N",
    "71_N","72_N","72_N","72_N","72_N","72_N","72_N","72_N","75_N","75_N","75_N","75_N","WT1_N5","WT2_N3")
Ts=c("2_T2","2_T3","4_T","5_T","16_T","19_T1","19_T3","19_T4","19_T5","47_T1","47_T4","47_T5","47_T6","47_T8",
    "49_T1","49_T3","50_T","65_T1","65_T3","66_T",
     "132_T1","151_T4","17_T4","18_T3","18_T4","142_T1","142_T2","142_T5-4","148_T1","148_T3",
     "68_P1","68_P2","68_P3","68_P4","68_P5-1","68_P5-2","68_P5-3","68_P5-4","68_P5-5",#"49_T2",
    "71_P3","72_P1","72_P2","72_P4","72_P5","72_P6","72_P8","72_P9","75_P3","75_P5","75_P8","75_P11","WT1_N7","WT2_N5")


for(my.sample in 1:length(Ns)){
    message("Merging ",Ns[my.sample], " and ", Ts[my.sample],"...")
    N <- read.delim(paste0(INDIR,Ns[my.sample],"_filtered_re.phy"),header = F, sep = " ",
                    colClasses=c("character","character"),skip = 1,
                    col.names = c("cellID","bi"))
    T <- read.delim(paste0(INDIR,Ts[my.sample],"_filtered_re.phy"),header = F, sep = " ",
                    colClasses=c("character","character"),skip = 2,
                    col.names = c("cellID","bi"))
    x <- rbind(N,T)
    ## 
    numSite=nchar(x$bi[1])
    ##
    colnames(x) <- c(as.character(nrow(x)), as.character(numSite))
    message("Number of Ns ",nrow(N), " and number of Ts ",nrow(T))
    write.table(x, file=paste0(OUTDIR, Ts[my.sample], "_filtered_re.phy"),sep=" ", row.names = F, quote = F)
}
```
